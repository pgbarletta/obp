```julia
using DataFrames
using Gadfly
using Compose
set_default_plot_size(30cm, 12cm)
```

## wt_apo

```julia
dir_wt_apo_pdt = "/home/german/labo/17/obp/run/wt_apo"
# RMSF
wt_apo_pdt_rmsf = readdlm("$dir_wt_apo_pdt/data/rmsf_wt_apo");
df_wt_apo_PDT_RMSF = DataFrame(RMSF=wt_apo_pdt_rmsf[:, 2], Frame=wt_apo_pdt_rmsf[:, 1]);

pdf_wt_apo_PDT_RMSF = plot(df_wt_apo_PDT_RMSF, Geom.line, x="Frame", y="RMSF",
Guide.title("Production"),
Guide.annotation(compose(compose(context(), text(70000, 3.5, "frame = 10ps")), fontsize(15pt), fill("black"))),
Theme(minor_label_font_size = 10pt, major_label_font_size = 15pt,
key_title_font_size=20pt, key_label_font_size=20pt))
```
